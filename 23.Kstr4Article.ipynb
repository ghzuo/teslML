# Read data

```python
## read text by input
# text = """算法应用于一些实际模拟体系。第四章，我们介绍了如何将其
# 应用于布朗粒子的运动、丙氨酸十二肽Ala12和Trp-cage蛋白的构象动力学的研究。
# 在对布朗粒子的研究中，我们使用布朗动力学模拟多条初始位置不同的轨迹，
# 然后利用EspcTM方法从模拟轨迹中得到了三种状态。"""

## read text from local file
# text = open("news.txt", 'r' ,encoding='utf-8').read()

## read data from url
text = kst.readurl("https://new.qq.com/omn/20220115/20220115A0651M00.html")

## delete space, Engnish letter, number, and punctuation from string
text = kst.clean(text)

## output information
print("Length of text: ", len(text))
```

# Cut Words by Jieba

```python
## get word data by jieba
jdata = kst.getwdata(jieba.lcut(text,cut_all=True, HMM=True))
```

```python
### plot data
kst.wcloudplot(jdata, nout)
```

```python
jdata[0:nout]
```

# Cut Words with Length

```python
## get word data by k-lenght cut
kmin = 1
kmax = 3
kdata = kst.getwdata(kst.eqcut(text, kmin, kmax))
```

```python
### plot data
kst.wcloudplot(kdata)
```
