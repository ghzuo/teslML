```python
# 常用数据处理包
import numpy as np        # 矩阵基本处理 ndarray
import scipy as sp        # 线性代数
import pandas as pd       # 数据整理 Serial, DataFrame
import matplotlib.pyplot as plt   # 画图工具
import seaborn as sns  # 画图，seaborn基于matplotlib, 其设置仍旧有效
```

```python
%matplotlib inline 
sns.set(font="DejaVu Sans", font_scale=1.8,
           rc={'figure.figsize': (10, 10),
            'lines.markersize': 15,
            "animation.embed_limit": 100})
```

# 基本参数

```python
nskip = 500
nstep = 500
```

# 抛物线映射

```python
def x2func(m, x):
    return 1 - m*x*x

zz = np.arange(-1,1,0.005)
yy = np.zeros(len(zz))
mu = np.arange(0,2,0.005)
zlist = np.repeat(zz, nstep)
mlist = np.repeat(mu, nstep)
ylist = np.zeros(len(zlist))
xlist = np.zeros(len(mlist))
```

```python
index = 0
m = 1
for z in zz:
    yy[index] = x2func(m, z)
    index = index + 1    
plt.plot(zz, yy, '-k', ms=0.2)
plt.show()
```

```python
index = 0
m = 1
for z in zz:
    yy[index] = x2func(m, z)
    index = index + 1    
plt.plot(zz, yy, '-k', ms=0.2)

for z in [i/20 for i in range(-20,1,1)]:
    plt.vlines(z, 0, x2func(m, z), color='red', linestyles = "dashed") 
    plt.hlines(x2func(m, z), -1, z, color='red', linestyles = "dashed") 
plt.show()
```

```python
index = 0
for m in mu :
    x = 0.5
    for i in np.arange(1, nskip, 1):
        x = x2func(m,x)
    for i in np.arange(1, nstep, 1):
        x = x2func(m,x)
        xlist[index] = x
        index = index + 1
plt.plot(mlist, xlist, 'ok', ms=0.2)
plt.show()
```

# 三角函数

```python
def x2func(m, x):
    return m*np.cos(np.pi*x)

zz = np.arange(-1,1,0.005)
yy = np.zeros(len(zz))
mu = np.arange(0,2,0.005)
zlist = np.repeat(zz, nstep)
mlist = np.repeat(mu, nstep)
ylist = np.zeros(len(zlist))
xlist = np.zeros(len(mlist))
```

```python
index = 0
m = 1.0
for z in zz:
    yy[index] = x2func(m, z)
    index = index + 1    
plt.plot(zz, yy, '-k', ms=0.2)
plt.show()
```

```python
index = 0
m = 1
for z in zz:
    yy[index] = x2func(m, z)
    index = index + 1    
plt.plot(zz, yy, '-k', ms=0.2)

for z in [i/20 for i in range(-20,1,1)]:
    plt.vlines(z, -1, x2func(m, z), color='red', linestyles = "dashed") 
    plt.hlines(x2func(m, z), -1, z, color='red', linestyles = "dashed") 
plt.show()
```

```python
index = 0
for m in mu :
    x = 0.5
    for i in np.arange(1, nskip, 1):
        x = x2func(m,x)
    for i in np.arange(1, nstep, 1):
        x = x2func(m,x)
        xlist[index] = x
        index = index + 1
plt.plot(mlist, xlist, 'ok', ms=0.2)
plt.show()
```
